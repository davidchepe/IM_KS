```python
import pandas as pd
from os import listdir, rename

from os.path import isfile, join
```

```python
NO EJECUTAR


xlsx = open('Combined_data.xlsx','wb')
result = pd.DataFrame()
for f in listdir():
    if f[-4:]=='xlsx' and f.startswith('Datos_Iroman'):
        new = pd.read_excel(f)
        result = pd.concat([result, new])
        result.to_excel('Combined_data.xlsx')
    print('finish', f)
xlsx.close()
```

```python
csv = open('Combined_data.csv','wb')
result = pd.DataFrame()
for f in listdir():
    if f[-4:]=='xlsx' and f.startswith('Datos_Iroman'):
        new = pd.read_excel(f)
        result = pd.concat([result, new])
        result.to_csv('Combined_data.csv')
    print('finish', f)
csv.close()
print('finish')
```
