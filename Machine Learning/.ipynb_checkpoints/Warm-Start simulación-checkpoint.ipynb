```python
import os
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
plt.style.use('seaborn')
import numpy as np
```

### Cargamos los datos

```python
#path = '../Entrenamientos/result_tramos.xlsx'
path = '../Entrenamientos/Procesado_Tramos.xlsx'
```

```python
df = pd.read_excel(path)
```

```python
df.describe()
```

```python
df = df[df["viento_aparente"]!=-np.inf]
df = df[df["porc"]!=np.inf]
df["puntos_en_tramo"]= df["tramo_final"]-df["tramo_inicial"]
df = df[df["puntos_en_tramo"]>10]
df = df[df["dist"]>400]
```

```python
df.describe()
```

```python
df.head()
```

```python
df_prueba = df.copy()
df_prueba["suma_prod"] = df_prueba["pwr"]*df_prueba["time_seg"]
```

```python
vatios_medio = df_prueba["suma_prod"].sum()/df_prueba["time_seg"].sum()
vatios_medio
```

```python
df_prueba["diff_vatios"] = df_prueba["pwr"]/vatios_medio

df_prueba
```

```python
from sklearn import datasets, linear_model
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
df_lm = df[['pwr','porc']]
df_lm


df_lm_prueba = df_prueba[['diff_vatios','porc']]
df_lm_prueba
```

```python
plt.scatter(df_lm['porc'],df_lm['pwr'])
```

```python
plt.scatter(df_lm_prueba['porc'],df_lm_prueba['diff_vatios'])
```

```python
df_lm_x = df_lm[['porc']]
df_lm_y = df_lm[['pwr']]


df_lm_x_prueba = df_lm_prueba[['porc']]
df_lm_y_prueba = df_lm_prueba[['diff_vatios']]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df_lm_x, df_lm_y, test_size=0.30, random_state=42)
X_train_prueba, X_test_prueba, Y_train_prueba, Y_test_prueba = train_test_split(df_lm_x_prueba, df_lm_y_prueba, test_size=0.30, random_state=42)
```

```python
X_train_prueba
```

### Modelo Lineal

```python
from sklearn import linear_model
regr = linear_model.LinearRegression(fit_intercept=True)
regr.fit(X_train,Y_train)

regr_prueba = linear_model.LinearRegression(fit_intercept=True)
regr_prueba.fit(X_train_prueba,Y_train_prueba)

Y_pred = regr.predict(X_test)

Y_pred_prueba = regr.predict(X_test_prueba)
```

```python
print(metrics.mean_squared_error(Y_test,Y_pred))
print(metrics.explained_variance_score(Y_test, Y_pred))
```

```python
print(metrics.mean_squared_error(Y_test_prueba,Y_pred_prueba))
print(metrics.explained_variance_score(Y_test_prueba, Y_pred_prueba))
```
